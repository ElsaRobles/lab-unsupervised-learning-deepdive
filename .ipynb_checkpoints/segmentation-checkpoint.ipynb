```python
stats.dropna(inplace=True)
```

```python
stats.info()
```

```python
stats_dum = pd.get_dummies(stats.gender, drop_first=True)
stats_analysis = pd.concat([stats, stats_dum], axis=1)
stats_analysis.drop('gender', axis=1, inplace =True)
```

## 4. Modeling

Now that your data is ready, apply one (or more) unsupervised learning model to find the groups of teens that exist:
* What model did you choose?
* What is the optimal number of clusters?
* Once your clusters are defined, how do you interpret them? - What type of teens there are? What are the differences between them?

```python
#I will perfome KMeans y DBSSCAN.
#With the Elbow Visualizer we can consider that the best number of clusters is 7. 

model = KMeans()
visualizer = KElbowVisualizer(model, k=(4,12))
visualizer.fit(stats_analysis)
visualizer.poof();
```

```python
# Your code here:

#I considered to scale de data to work with the outliers. 

scaler = StandardScaler() #To initialize the scaler

#.fit= compute the mean and std to be used for later scaling.
#.fit_transform Fit to data, then transform it.
#Passing the DataFrame thorug the scaler and transformed it.

stats_scale = pd.DataFrame(scaler.fit_transform(stats_analysis))
#Assingning the transformation to the new variable
```

```python
stats_scale
```

```python
#I started with  KMeans. 
kmeans = KMeans(n_clusters=7) #Model initialization 

#Applying the model to the scaled data
stats_clustered = kmeans.fit(stats_scale) 

stats_analysis['labels']= stats_clustered.labels_ 
```

```python
len(stats_clustered.labels_ ) 

#We haven't missed observations
```

```python
# Your code here
dbscan_model = DBSCAN(eps=0.7)  #Model initialization 

#Applying the model to the scaled data
dbscan_model.fit(stats_scale)

stats_analysis['labels_DBSCAN'] = dbscan_model.labels_
```

```python
print(len(dbscan_model.labels_))
#We haven't missed observations
```

```python
f, axes = plt.subplots(1, 2, figsize=(12,8)) 


sns.scatterplot(y="die", x= "gradyear", hue="labels", data=stats_analysis, ax=axes[0], palette = "Greens");
sns.scatterplot(y="die", x= "gradyear", hue="labels_DBSCAN", data=stats_analysis, ax=axes[1], palette = "Greens");
```

```python
stats_analysis.groupby('labels').mean()
```

```python
stats_analysis.groupby('labels_DBSCAN').mean()
```
